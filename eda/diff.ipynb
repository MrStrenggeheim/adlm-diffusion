```python

img_np = np.array(img)
img_np = img_np / 255
plt.imshow(img_np, cmap='gray')
```

```python
noise_np = np.random.normal(0, .5, img_np.shape)
noise_level = .5
noised_img_np = img_np + noise_level * noise_np
noised_img_np = (noised_img_np - noised_img_np.min()) 
noised_img_np /= noised_img_np.max()
print(noised_img_np.min(),noised_img_np.max())

# plt.imshow(noise_np, cmap='gray')
plt.imshow(noised_img_np, cmap='gray', vmin=0, vmax=1)
Image.fromarray((noised_img_np * 255).astype(np.uint8)).save("noised_img.png")
```
